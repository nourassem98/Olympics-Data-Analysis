```python
# Replacing Any String NAN to np.nan
df_olympic_history.replace('NA', np.nan, inplace=True)
df_olympic_history.replace('Missing', np.nan, inplace=True)
df_olympic_history.replace('NaN', np.nan, inplace=True)
```

```python
# Box plot of the Age col. to perview outliers
sns.boxplot(x='Age', data=df_olympic_history)
plt.title('Boxplot of Age')
plt.show()
```

```python
# Histogram plot of the Age col. to perview outliers

bins = [10,15,20,25,30,35,40,45,50,55,60,65,70,75]#the intervals (bins) of the histogram
fig, axs_1 = plt.subplots(1,1,dpi=100)
axs_1.set_title('Age Histogram')
axs_1.set_xlabel('Age')
axs_1.set_xticks(bins)
# axs_3.set_xlim([35,90]) #zoom in, or show me a certain area on the graph 
axs_1.set_ylabel('Count')
axs_1.hist(df_olympic_history['Age'],width=5,bins=bins)
```

```python
# Calculating IQR for Age 
Q1=df_olympic_history.Age.quantile(0.25)
Q3=df_olympic_history.Age.quantile(0.75)
IQR=Q3-Q1
lowqe_bound=Q1 - 1.5 * IQR
upper_bound=Q3 + 1.5 * IQR
print(lowqe_bound,upper_bound)
```

```python
# Dropping Age Outliers from the dataSet 
df_olympic_history.drop(df_olympic_history[df_olympic_history['Age'] < lowqe_bound].index, inplace = True)
df_olympic_history.drop(df_olympic_history[df_olympic_history['Age'] > upper_bound].index, inplace = True)

df_olympic_history.shape
```

```python
# Calculating the mean of Age According to the gender in each sport
df_olympic_history.groupby(['Sex','Sport'])['Age'].mean()
```

```python
df_olympic_history['Age'] = df_olympic_history.groupby(['Sex','Sport'])['Age'].transform(lambda x: x.fillna(x.mean()))
```
