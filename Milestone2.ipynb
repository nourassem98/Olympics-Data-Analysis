Binning the BMI so it looks tidier in visualization step.

```python
BMI_group = [] # define array structure
for BMI in df_olympic_history["BMI"]:
    if BMI<13:
        BMI_group.append("<13")
    elif 13 <= BMI <= 16:
        BMI_group.append("13-16")
    elif 16 < BMI <= 19:
        BMI_group.append("16-19")
    elif 19 < BMI <= 21:
        BMI_group.append("19-21")
    elif 21 < BMI <= 24:
        BMI_group.append("21-24")
    elif 24 < BMI <= 27:
        BMI_group.append("24-27")
    elif 27 < BMI <= 30:
        BMI_group.append("27-30")
    elif 30 < BMI <= 33:
        BMI_group.append("30-33")
    elif 33 < BMI <= 36:
        BMI_group.append("33-36")
    elif 36 < BMI <= 39:
        BMI_group.append("36-39")
       
    else:
        BMI_group.append("40 and over")
        
# Copy dataframe to keep original 
new_income_df = df_olympic_history.copy()
new_income_df["BMI"] = BMI_group
new_income_df
```

```python
male_medal=new_income_df[new_income_df['Sex']=='M'].groupby(['BMI'])['Medal'].count()
male1= (male_medal.sort_values(ascending=False))
male1.head(10)
male1=male_medal.sort_values(axis=0, ascending=False, inplace=False)
```

```python
m=male1.to_frame()
m=m.sort_values(by=['BMI'])
m.index
```

```python
female_medal=new_income_df[new_income_df['Sex']=='F'].groupby(['BMI'])['Medal'].count()
female1= (female_medal.sort_values(ascending=False))
female1.head(10)
female1=female_medal.sort_values(axis=0, ascending=False, inplace=False)
```

```python
f=female1.to_frame()
f=f.sort_values(by=['BMI'])
f.index
```
